```
import pandas as pd
import json
import zipfile
```

```
!wget https://github.com/TUB-NLP-OpenData/climate_change/releases/download/0.1.2/train_test.zip
```

```
archive = zipfile.ZipFile('train_test.zip', 'r')
file_data = archive.open('train_test.json')
```

```
df = pd.DataFrame(json.load(file_data))
df = df[df["tweet_class"] != "unknown"]
```

```
df.head()
```

```
ages = []
genders = []
for index, row in df.iterrows():
  if len(row["faces_predict"]) >= 1:
    age = row["faces_predict"][0]["age"]
    gender = row["faces_predict"][0]["gender"]
    ages.append(age)
    if gender > 0.5:
      gender = "F"
    else:
      gender = "M"
    genders.append(gender)
  else:
    ages.append("")
    genders.append("")

df["gender"] = genders
df["age"] = ages
```

```
import altair as alt
alt.data_transformers.enable('default', max_rows=None)
```

```
points = alt.Chart(df).mark_point().encode(
  x=alt.X('mean(age):Q', title="Age"),
  y=alt.Y('gender:N', title="Gender"),
  color='tweet_class',
)

points
```

```
points = alt.Chart(df).mark_point().encode(
  x=alt.X('age:Q', title="Age"),
  y=alt.Y('gender:N', title="Gender"),
  color='tweet_class',
)

points
```
